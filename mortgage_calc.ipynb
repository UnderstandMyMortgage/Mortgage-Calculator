```python
import math
import pandas as pd
```

```python
# Get user inputs
property_value = 500000 #float(input("Enter the total property value: "))
deposit_amount = 100000 #float(input("Enter the deposit amount: "))
interest_rate = 0.05 #float(input("Enter the interest rate (as a decimal): "))
loan_term = 20 #int(input("Enter the loan term (in years): "))
```

```python
# Calculate the loan amount
loan_amount = property_value - deposit_amount
# Calculate the monthly interest rate
monthly_interest_rate = interest_rate / 12
# Calculate the number of monthly payments
num_payments = loan_term * 12
# Calculate the monthly repayment
monthly_repayment = (loan_amount * monthly_interest_rate) / (1 - math.pow(1 + monthly_interest_rate, -num_payments))
# Print the monthly repayment
print("Monthly repayment: $%.2f" % monthly_repayment)
# Calculate the total amount repaid
total_amount_repaid = monthly_repayment * num_payments
# Calculate the total interest paid
total_interest_paid = total_amount_repaid - loan_amount
```

```python
# Calculate the amount of each payment that goes towards interest and repaying the principle
balance = loan_amount
for i in range(num_payments):
    interest_paid = balance * monthly_interest_rate
    principle_paid = monthly_repayment - interest_paid
    balance -= principle_paid
    print("Payment %2d  |  Interest: $%8.2f  |  Principle: $%8.2f  |  Remaining Balance: $%8.2f" % (i+1, interest_paid, principle_paid, balance))

print("Total interest paid:         $%.2f" % total_interest_paid)
print("Total amount repaid:         $%.2f" % total_amount_repaid)
```

```python
payments_df = pd.DataFrame(columns=["Payment", "Interest", "Principle", "Balance"])

balance = loan_amount
for i in range(num_payments):
    interest_paid = balance * monthly_interest_rate
    principle_paid = monthly_repayment - interest_paid
    balance -= principle_paid
    payments_df.loc[i] = [i+1, interest_paid, principle_paid, balance]
```

```python
filename = "mortgage_payments_" + str(property_value) + "value_" + str(loan_amount) + "loan_amount_" + str(interest_rate*100) + "pctinterest" + ".csv"
payments_df.to_csv(filename, index=False)
```

```python
property_value = 500000 # float(input("Enter the total property value: "))
deposit_amount = 100000 # float(input("Enter the deposit amount: "))
intro_rate = 0.02 # float(input("Enter the introductory interest rate (as a decimal): "))
standard_rate = 0.06 # float(input("Enter the standard interest rate (as a decimal): "))
intro_term = 3 # int(input("Enter the introductory term (in years): "))
loan_term = 20 # int(input("Enter the loan term (in years): "))
```

```python
# Calculate the loan amount
loan_amount = property_value - deposit_amount

# Calculate the number of monthly payments
num_payments = loan_term * 12

# Create an empty list to store the payment information
payments = []
total_interest_paid = 0
total_amount_repaid = 0

# Calculate the monthly repayment
balance = loan_amount
for i in range(num_payments):
    # Calculate the monthly interest rate
    if i < intro_term * 12:
        monthly_interest_rate = intro_rate / 12
    else:
        monthly_interest_rate = standard_rate / 12
    
    # Calculate the monthly repayment
    monthly_repayment = (monthly_interest_rate * balance) / (1 - math.pow(1 + monthly_interest_rate, -num_payments + i))
    total_amount_repaid += monthly_repayment
    
    # Calculate the amount of each payment that goes towards interest and repaying the principle
    interest_paid = balance * monthly_interest_rate
    total_interest_paid += interest_paid
    principle_paid = monthly_repayment - interest_paid
    balance -= principle_paid
    
    # Add the payment information to the list
    payments.append((i+1, interest_paid, principle_paid, balance))

# Print the payment information
for payment in payments:
    print("Payment %2d - Interest: $%8.2f, Principle: $%8.2f, Balance: $%10.2f" % payment)

# Print some summary statistics about the mortgage
print("Total amount repaid: $%0.2f" % total_amount_repaid)
print("Total interest paid: $%0.2f" % total_interest_paid)
```
